**Steam Web API**

```python
import requests

api_key = "25FF15A5************************" # API kľúč je súkromný
steam_id = "76561198271641008"
app_id = "730" # CS
```

```python
# Funkcia na spracovanie odpovedí a vypísanie údajov
def fetch_data(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Error fetching data:", response.status_code)
        return None
```

```python
# Získanie údajov o hráčovi
url_player = f"http://api.steampowered.com/ISteamUser/GetPlayerSummaries/v0002/?key={api_key}&steamids={steam_id}"

player_data = fetch_data(url_player)
if player_data and 'response' in player_data and 'players' in player_data['response']:
    player_info = player_data['response']['players'][0]
    print(f"Username: {player_info['personaname']}")
    print(f"Profile URL: {player_info['profileurl']}")
    print(f"Avatar: {player_info['avatarfull']}")
    print(f"Country Code: {player_info.get('loccountrycode', 'Not provided')}")
```

```python
# URL - získanie štatistík pre danú hru
url_stats = f"http://api.steampowered.com/ISteamUserStats/GetUserStatsForGame/v0002/?appid={app_id}&key={api_key}&steamid={steam_id}"

stats_data = fetch_data(url_stats)
if stats_data and 'playerstats' in stats_data:
    stats = stats_data['playerstats']
    for stat in stats['stats']:
        print(f"{stat['name']}: {stat['value']}")
```

```python
# URL na získanie achievementov hráča
url_achievements = f"http://api.steampowered.com/ISteamUserStats/GetPlayerAchievements/v0001/?appid={578080}&key={api_key}&steamid={steam_id}"
achievements_data = fetch_data(url_achievements)

# URL na získanie schémy hry (názvy a popisy achievementov)
url_schema = f"http://api.steampowered.com/ISteamUserStats/GetSchemaForGame/v2/?key={api_key}&appid={578080}"
schema_data = fetch_data(url_schema)

# Mapovanie ID na názvy a popisy achievementov
achievement_info = {}
if schema_data and 'game' in schema_data and 'availableGameStats' in schema_data['game']:
    for achievement in schema_data['game']['availableGameStats'].get('achievements', []):
        achievement_info[achievement['name']] = {
            'displayName': achievement.get('displayName', 'Unknown'),
            'description': achievement.get('description', 'No description')
        }

# Výpis achievementov hráča so skutočnými názvami a popismi
if achievements_data and 'playerstats' in achievements_data:
    achievements = achievements_data['playerstats'].get('achievements', [])
    print("Achievements:")
    for achievement in achievements:
        name = achievement.get('apiname', 'Unknown')
        achieved = "Unlocked" if achievement.get('achieved', 0) else "Not Unlocked"

        # Podrobnosti
        display_name = achievement_info.get(name, {}).get('displayName', name)
        description = achievement_info.get(name, {}).get('description', 'No description')

        print(f"- {display_name}: {achieved}")
        print(f"  Description: {description}")
```

```python
# URL - získanie globálnych percent úspechov pre danú hru
url_global_achievements = f"http://api.steampowered.com/ISteamUserStats/GetGlobalAchievementPercentagesForApp/v0002/?gameid={578080}&format=json"
global_achievements_data = fetch_data(url_global_achievements)

# Získanie globálnych percentuálnych úspechov pre každý achievement
global_achievement_percentages = {}
if global_achievements_data and 'achievementpercentages' in global_achievements_data:
    for achievement in global_achievements_data['achievementpercentages']['achievements']:
        global_achievement_percentages[achievement['name']] = achievement['percent']

# Výpis achievementov hráča so skutočnými názvami, popismi a globálnymi percentami
if achievements_data and 'playerstats' in achievements_data:
    achievements = achievements_data['playerstats'].get('achievements', [])
    print("Achievements:")
    for achievement in achievements:
        name = achievement.get('apiname', 'Unknown')

        # podrobnosti
        display_name = achievement_info.get(name, {}).get('displayName', name)
        description = achievement_info.get(name, {}).get('description', 'No description')

        # Globálne percento úspešnosti pre daný achievement
        global_percentage = global_achievement_percentages.get(name, 'N/A')

        print(f"- {display_name}: ")
        print(f"  Description: {description}")
        print(f"  Achievement získalo {global_percentage}% hráčov.")
```

```python
# URL - získanie noviniek pre danú hru
url_news = f"http://api.steampowered.com/ISteamNews/GetNewsForApp/v0002/?appid={app_id}&count=3&feeds=steamcommunity&key={api_key}"

news_data = fetch_data(url_news)
if news_data and 'appnews' in news_data:
    for news_item in news_data['appnews']['newsitems']:
        print(f"Title: {news_item['title']}")
        print(f"Date: {news_item['date']}")
        print(f"URL: {news_item['url']}\n")
```

```python
# Získanie vlastnených hier
url_owned_games = f"http://api.steampowered.com/IPlayerService/GetOwnedGames/v0001/?key={api_key}&steamid={steam_id}&include_appinfo=true&include_played_free_games=true"

owned_games_data = fetch_data(url_owned_games)
if owned_games_data and 'response' in owned_games_data and 'games' in owned_games_data['response']:
    print("Owned Games:")
    for game in owned_games_data['response']['games']:
        print(f"- {game['name']} (Playtime: {game['playtime_forever']} minutes)")
```

```python
# Získanie nedávno hraných hier
url_recently_played = f"http://api.steampowered.com/IPlayerService/GetRecentlyPlayedGames/v0001/?key={api_key}&steamid={steam_id}&format=json"

recently_played_data = fetch_data(url_recently_played)
if recently_played_data and 'response' in recently_played_data and 'games' in recently_played_data['response']:
    print("Recently Played Games:")
    for game in recently_played_data['response']['games']:
        print(f"- {game['name']} (Playtime: {game['playtime_2weeks']} minutes)")
```
